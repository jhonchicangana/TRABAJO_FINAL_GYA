Se transforma el dataFrame en una tabla sqlite3 llamada proveedor

```python
# Ejecuta una consulta SQL para obtener información sobre las columnas de la tabla
conn = sql.connect("db_proveedores.db")
cursor = conn.cursor()
cursor.execute(f"PRAGMA table_info(proveedor)")
# Recupera los resultados de la consulta
columnas = cursor.fetchall()
# Imprime el nombre y otros detalles de las columnas
for columna in columnas:
    print(columna[1])  # La segunda posición contiene el nombre de la columna
# Cierra la conexión a la base de datos
conn.close()
```

Se listan los campos que contiene la tabla proveedor.  

```python
conn = sql.connect("db_proveedores.db")
cursor = conn.cursor()
query = f"SELECT * FROM proveedor LIMIT 20;"
cursor.execute(query)
conn.commit()
resultados = cursor.fetchall()
for fila in resultados:
  print(fila)
print(resultados)
conn.close()
```
